```python
def get_attributes(json): 
    tuples = []
    names = []
    try:
        for x in range(len(json)):
            name = json['object'][x]['name']
            xmin = int(json['object'][x]['bndbox']['xmin'])
            ymin = int(json['object'][x]['bndbox']['ymin'])
            xmax = int(json['object'][x]['bndbox']['xmax'])
            ymax = int(json['object'][x]['bndbox']['ymax'])
            filename = json['filename']
            
            attributes = [ymin, xmin, ymax, xmax]
            
            tuples.append(attributes)
            names.append(name)
        
    except IndexError:
        pass
    #print(filename)
    return ast.literal_eval(str(tuple([[tuples] + [names]])).replace('(','').replace(',)',''))
```
